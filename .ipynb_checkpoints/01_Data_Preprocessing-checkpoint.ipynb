### Analyzing Different Findings

```python
findings = set()
for finding in df_cxr_data['Finding Labels'].unique():
    print(finding.split('|'))
    findings.update(finding.split('|'))
```

```python
print(f'There are total {len(findings)} findings: The list of findings are {findings}')
```

```python
for finding in findings:
    df_cxr_data[finding] = df_cxr_data['Finding Labels'].map(lambda x: 1 if finding in x else 0)
```

### Age Analysis

```python
df_cxr_data['Patient Age'].unique()
```

```python
#Dealing with outliers(Ages pver 100)

age_outliers = df_cxr_data[df_cxr_data['Patient Age']>100]
print(len(age_outliers))
```

```python
# Fixing the Age Outliers
df_cxr_data['Patient Age'] = df_cxr_data.apply(lambda x: 101 if x['Patient Age'] > 100 else x['Patient Age'], axis=1)
```

```python
df_cxr_data['Patient Age'].plot(kind='hist', bins=10)
plt.xlabel("Age")
plt.ylabel("Number of people")
plt.title('Age Distribution of Patients')
plt.tight_layout()
plt.show()
```

### Image Analysis

```python
def show_image_hist(img_path):
    
    f=plt.figure(figsize= (20,8))
    s1 = f.add_subplot(1, 2, 1)
    s1.set_title("Patient's Image")
    img = cv2.imread(img_path)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    
    s2 = f.add_subplot(1,2, 2)
    s2.set_title('Intensity Distribution')
    plt.hist(img.ravel(), bins = 256)

    

    plt.show()
```

```python
show_image_hist(DATA_PATH+'/images_001/images/00000003_005.png')
```

```python
xray_samples = list(DATA_PATH+'/images_001/images/' + df_cxr_data['Image Index'][0:5])
xray_samples
```

```python
for xray_sample in xray_samples:
    show_image_hist(xray_sample)
```
